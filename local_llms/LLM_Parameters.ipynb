```python
# Install transformers (lightweight installation)
!pip install -q transformers

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, set_seed

# Use a tiny model for demonstration
model_name = "sshleifer/tiny-gpt2"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()

set_seed(42)

def generate_text(prompt, temperature=1.0, top_k=50, top_p=1.0, max_length=50):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")
    with torch.no_grad():
        output = model.generate(
            input_ids,
            do_sample=True,
            temperature=temperature,
            top_k=top_k,
            top_p=top_p,
            max_length=max_length,
        )
    return tokenizer.decode(output[0], skip_special_tokens=True)

prompt = "The future of AI is"

print("Low Temperature (0.3):", generate_text(prompt, temperature=0.3))
print("High Temperature (1.5):", generate_text(prompt, temperature=1.5))
print("Top-k = 5:", generate_text(prompt, top_k=5))
print("Top-p = 0.7:", generate_text(prompt, top_p=0.7))
```
