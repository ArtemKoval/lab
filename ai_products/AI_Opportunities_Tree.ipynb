```python
!pip install openai matplotlib scikit-learn --quiet

import openai
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from google.colab import userdata

client = openai.OpenAI(api_key=userdata.get('OPEN_API_KEY'))

def ask_chatgpt(prompt, max_tokens=150, temperature=0.7):
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}],
        temperature=temperature,
        max_tokens=max_tokens,
    )
    return response.choices[0].message.content.strip()

support_ticket = """
Hi, I’m trying to reset my password but the email link isn’t working. I’ve tried twice now and it keeps saying “expired.”
Please help me regain access to my account as soon as possible.
"""
summary_prompt = f"Summarize this support ticket in one sentence:\n\n{support_ticket}"
print("Summary:", ask_chatgpt(summary_prompt))

marketing_prompt = "Write 3 email subject lines for a winter clothing sale targeted at eco-conscious customers."
print("Marketing Ideas:", ask_chatgpt(marketing_prompt))

user_tags = ["AI in healthcare", "machine learning trends", "data privacy"]
articles = [
    "Blockchain and Security",
    "Healthcare AI Applications",
    "Latest in Machine Learning",
    "Privacy Regulation Explained",
    "Home Gardening Tips"
]
recommendation_prompt = f"""
Based on these user interests: {', '.join(user_tags)},
recommend 3 relevant article titles from this list: {', '.join(articles)}.
"""
print("Recommended Articles:", ask_chatgpt(recommendation_prompt))

innovation_prompt = "List 5 startup ideas that use AI to improve education."
print("Startup Ideas:", ask_chatgpt(innovation_prompt))

def predict_next_date(dates):
    diffs = np.diff(dates)
    avg_diff = int(np.mean(diffs))
    return dates[-1] + avg_diff

dates = [1, 8, 15, 22]
print("Predicted next meeting date (day of month):", predict_next_date(dates))

sentiment_prompt = "Analyze the sentiment of this feedback: 'I love the new features in your app. It really makes things easier!'"
print("Sentiment:", ask_chatgpt(sentiment_prompt))

benefit_areas = ["Automation", "Augmentation", "Personalization", "Innovation", "Convenience", "Emotional"]
impact_scores = [8, 7, 6, 9, 5, 6]

plt.figure(figsize=(8, 5))
plt.barh(benefit_areas, impact_scores, color="steelblue")
plt.xlabel("Perceived Business Impact")
plt.title("AI Benefit Areas and Their Potential")
plt.grid(axis="x")
plt.show()
```
