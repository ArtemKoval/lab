```python
# First install the latest openai package in Colab
# !pip install openai

import os
from openai import OpenAI
from google.colab import userdata

# Set your OpenAI API key here or use an environment variable
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API_KEY')

# Initialize the OpenAI client
client = OpenAI()

def ask_chatgpt(prompt):
    # Use the chat.completions.create() method with correct syntax
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant for IT tasks."},
            {"role": "user", "content": prompt}
        ],
        max_tokens=200,
        temperature=0.5
    )
    return response.choices[0].message.content.strip()

# Example 1: Generate documentation
doc_prompt = "Write a technical document explaining how to configure SSH on Ubuntu Linux."
print("=== Generated Documentation ===")
print(ask_chatgpt(doc_prompt))
print("\n")

# Example 2: Write a Bash script
script_prompt = "Generate a Bash script to find and list all large files over 1GB in the /var/log directory."
print("=== Generated Script ===")
print(ask_chatgpt(script_prompt))
print("\n")

# Example 3: Log analysis explanation
log_sample = """
Jul 10 08:22:15 server sshd[1234]: Failed password for root from 192.168.1.10 port 22 ssh2
Jul 10 08:22:16 server sshd[1235]: Accepted password for admin from 192.1.1.1.20 port 23 ssh2
"""
log_prompt = f"Analyze the following log entries and explain what they mean:\n{log_sample}"
print("=== Log Analysis ===")
print(ask_chatgpt(log_prompt))
print("\n")

# Example 4: Performance review writing
review_prompt = "Write a performance review for an IT support technician using the SMART goal format."
print("=== Performance Review ===")
print(ask_chatgpt(review_prompt))
```
