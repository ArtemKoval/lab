```python
# Step 1: Install required libraries
!pip install transformers accelerate sentencepiece

# Step 2: Import necessary modules
import torch
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

# Step 3: Load model and tokenizer
model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Move to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Helper function to generate responses
def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True).to(device)
    outputs = model.generate(**inputs, max_new_tokens=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)

# Task 1: Text Classification
prompt = "Is this review positive or negative? 'I absolutely love this product! It works perfectly.'"
print("Classification:\n", generate_response(prompt))

# Task 2: Code Generation
prompt = "Write a Python function to reverse a string."
print("\nCode Generation:\n", generate_response(prompt))

# Task 3: Math Reasoning
prompt = "Solve 3x + 5 = 20 for x."
print("\nMath Solving:\n", generate_response(prompt))

# Task 4: Dialogue
prompt = "User: What's the capital of France?\nAssistant:"
print("\nDialogue:\n", generate_response(prompt))
```
