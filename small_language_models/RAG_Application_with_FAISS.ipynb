```python
# Step 1: Install required packages
!pip install faiss-cpu sentence-transformers pandas numpy

# Step 2: Import libraries
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss

# Step 3: Setup sample data
data = {
    "text": [
        "Spider-Man swings between buildings and shoots webs.",
        "Peter Parker is Spider-Man and he's a friendly neighborhood hero.",
        "Iron Man builds advanced suits and protects the world.",
        "Tony Stark is the genius behind Iron Man.",
        "Captain America fights for justice with his shield.",
        "Steve Rogers was enhanced to become Captain America."
    ],
    "label": [
        "spiderman", "spiderman",
        "ironman", "ironman",
        "captainamerica", "captainamerica"
    ]
}

df = pd.DataFrame(data)
print("Sample Data:")
print(df)
print("\n")

# Step 4: Generate sentence embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['text'].tolist())
embeddings = embeddings / np.linalg.norm(embeddings, axis=1, keepdims=True)

# Step 5: Build FAISS index
dimension = embeddings.shape[1]
index = faiss.IndexFlatIP(dimension)
index.add(np.array(embeddings).astype('float32'))

# Step 6: Perform similarity search
query_text = "Spider-Man uses his spider abilities to fight crime."
query_embedding = model.encode([query_text])
query_embedding = query_embedding / np.linalg.norm(query_embedding, axis=1, keepdims=True)

k = len(df)
distances, indices = index.search(np.array(query_embedding).astype('float32'), k)

# Display results
results = df.iloc[indices[0]]
results['distance'] = distances[0]
print("Search Results (sorted by similarity):")
print(results[['text', 'label', 'distance']])
```
